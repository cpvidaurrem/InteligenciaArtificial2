```python
import requests # librearia para hacer peticiones http
from bs4 import BeautifulSoup # libreria para hacer web scraping
```

```python
url = "https://listado.mercadolibre.com.ar/zapatilla-hombre#D[A:zapatilla%20hombre]" # URL de la página a scrapear

response = requests.get(url) # Hacemos la petición a la página

if response.status_code == 200: # Si la petición fue exitosa
    
    soup = BeautifulSoup(response.text, "html.parser") # Para convertir el HTML de la página en un objeto BeautifulSoup
    productos = soup.find_all("div", class_="andes-card poly-card poly-card--grid-card andes-card--flat andes-card--padding-0 andes-card--animated") # Buscamos todos los elementos con la clase "andes-card poly-card poly-card--grid-card andes-card--flat andes-card--padding-0 andes-card--animated"
    
    for producto in productos: # Iteramos sobre la lista de productos
        titulo = producto.find("h2", class_="poly-box poly-component__title") # Obtenemos el título del producto
        marca = producto.find("span", class_="poly-component__brand") # Obtenemos la marca del producto
        if titulo:
            print(titulo.text.strip()) # Imprimimos el título del producto
        if marca:
            print(marca.text.strip()) # Imprimimos la marca del producto
            
else:
    print("Error al cargar la pagina web, codigo de error: ", response.status_code)
```

*WEB SCRAPING CON SELENIUM*

Truco para hacer web scraping cuando la pagina que nos da error 403 por politicas que lo prohiben
- Indica que el acceso está prohibido debido a una regla de acceso específica definida en el servidor. Puede ser una restricción basada en direcciones IP, usuarios o cualquier otro criterio establecido por el administrador del servidor.

```python
from selenium import webdriver # Para poder hacer web scraping dinámico
from bs4 import BeautifulSoup # Para poder hacer web scraping dinámico

driver = webdriver.Chrome() # Ruta del driver de Chrome
url = "https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-bebidas-bebidas-sin-alcohol-jugos/_/N-11la5tu"
driver.get(url) # Hacemos la petición a la página

html = driver.page_source # Obtenemos el HTML de la página

soup = BeautifulSoup(html, "html.parser") # Para convertir el texto de la página en un objeto BeautifulSoup

productos = soup.find_all("div", class_="product_info_container") # Buscamos todos los elementos con la clase "product_info_container"
print (len(productos))

for producto in productos: # Iteramos sobre la lista de productos
        span_nombre = producto.find("span", class_="span_productName") # Obtenemos el nombre del producto
        div_nombre = span_nombre.find("div") # Obtenemos el nombre del producto
        print(div_nombre.text.strip()) # Imprimimos el nombre del producto
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://tecnologia.usfx.bo/" # URL de la página a scrapear

response = requests.get(url) # Hacemos la petición a la página

if response.status_code == 200: # Si la petición fue exitosa
    soup = BeautifulSoup(response.text, "html.parser") # Para convertir el texto de la página en un objeto BeautifulSoup
    licenciaturas = soup.find_all("li", class_="et_pb_menu_page_id-912 menu-item menu-item-type-custom menu-item-object-custom menu-item-has-children menu-item-912") # Buscamos todos los elementos con la clase "andes-card poly-card poly-card--grid-card andes-card--flat andes-card--padding-0 andes-card--animated"
    
    for licenciatura in licenciaturas: # Iteramos sobre la lista de licenciaturas
        licencitura = licenciatura.find("li") # Obtenemos la licenciatura
        print(licenciatura.text.strip()) # Imprimimos la licenciatura
    print ("Obtuvimos la pagina web")
            
else:
    print("Error al cargar la pagina web, codigo de error: ", response.status_code)
```
